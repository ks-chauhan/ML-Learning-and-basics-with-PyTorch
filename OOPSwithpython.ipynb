```python
class Car:
  def __init__(self,color,milage):
    self.color=color
    self.milage=milage
  def __str__(self):
    return f"The {self.color} Car has {self.milage} kilometers."
x=Car("blue",20000)
y=Car("yellow",1000000)
print(x,y)
```

```python
class Dog:
    species = "Canis familiaris"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} is {self.age} years old"

    def speak(self, sound):
        return f"{self.name} says {sound}"
```

```python
class GoldenRetriever(Dog):
  def speak(self,sound="Bark"):
    return super().speak(sound)
```

```python
print(GoldenRetriever("john",20).speak())
```
