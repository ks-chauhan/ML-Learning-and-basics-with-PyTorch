```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
```

```python
if (torch.cuda).is_available():
  device="cuda"
else:
  device="cpu"
```

```python
device
```

```python
torch.manual_seed(18)
torch.cuda.manual_seed(18)
```

```python
n_samples=10000
X,y=make_circles(n_samples,
                 noise=0.025,
                 random_state=18)
```

```python
y.shape
```

```python
pd.DataFrame({"X1" : X[:,0],
            "X2" : X[:,1],
            "labels" : y})
```

```python
plt.figure(figsize=(5,5))
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.RdYlBu)
```

```python
X_train_numpy,X_test_numpy,y_train_numpy,y_test_numpy=train_test_split(X,
                                               y,
                                               test_size=0.25,
                                               random_state=18)
```

```python
X_train,X_test,y_train,y_test=torch.tensor(X_train_numpy,dtype=torch.float32),torch.tensor(X_test_numpy,dtype=torch.float32),torch.tensor(y_train_numpy,dtype=torch.float32),torch.tensor(y_test_numpy,dtype=torch.float32)
```

```python
len(X_train)
```

```python
len(y_train)
```

```python
type(X_train)
```

```python
X_train,X_test,y_train,y_test=X_train.to(device),X_test.to(device),y_train.to(device),y_test.to(device)
```

```python
model_0=nn.Sequential(
    nn.Linear(in_features=2,out_features=8),
    nn.Linear(in_features=8,out_features=16),
    nn.ReLU(),
    nn.Linear(in_features=16,out_features=1)
)
```

```python
model_0.state_dict()
```

```python
loss_fn=torch.nn.BCEWithLogitsLoss()
```

```python
optimizer=torch.optim.SGD(params=model_0.parameters(),
                       lr=0.01)
```

```python
def accuracy(predicted,true):
  accu=torch.eq(torch.sigmoid(predicted).round(),true).sum().item()
  accuracy=(accu/len(predicted))*100
  return accuracy
```

```python
model_0.to(device)
```

```python
prediction=model_0(X_train)
prediction.shape
```

```python
y_train.shape
torch.sigmoid(prediction[:20]).round()
```

```python
accuracy(prediction.squeeze(),y_train)
```

```python
epochs=3000
for epoch in range(epochs):
  model_0.train()
  prediction=model_0(X_train)
  loss=loss_fn(prediction.squeeze(),y_train)
  accu=accuracy(prediction.squeeze(),y_train)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
  if(epoch%100==0):
    print(accu,loss)
```

```python
with torch.inference_mode():
  test_prediction=model_0(X_test)
test_prediction[:5]
```
